```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

# --- 1. Pauli-X Gate ---
qc_x = QuantumCircuit(1)
qc_x.x(0)  # Apply X gate to |0>
sv_x = Statevector(qc_x)

print("1. State after applying Pauli-X to |0>:")
display(sv_x.draw('latex')) 
# If latex drawing doesn't work in your environment, use: print(sv_x)

# --- 2. Pauli-Y Gate ---
qc_y = QuantumCircuit(1)
qc_y.y(0)  # Apply Y gate to |0>
sv_y = Statevector(qc_y)

print("\n2. State after applying Pauli-Y to |0>:")
display(sv_y.draw('latex'))

# --- 3. Pauli-Z Gate ---
qc_z = QuantumCircuit(1)
qc_z.x(0)  # First apply X to put it in the |1> state
qc_z.z(0)  # Then apply Z gate to see the phase flip
sv_z = Statevector(qc_z)

print("\n3. State after applying Pauli-Z to |1>:")
display(sv_z.draw('latex'))
```
