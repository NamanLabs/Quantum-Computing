```python
# Import the QuantumCircuit class from Qiskit
from qiskit import QuantumCircuit

# Step 1: Initialize a simple quantum circuit
# We are creating a circuit with 2 qubits and 2 classical bits
qc = QuantumCircuit(2, 2)

# Step 2: Apply some basic operations (gates)
# Apply a Pauli-X gate (NOT gate) to qubit 0
qc.x(0)

# Apply a Hadamard gate (Superposition) to qubit 1
qc.h(1)

# Step 3: Add measurement operations
# Measure qubit 0 into classical bit 0, and qubit 1 into classical bit 1
qc.measure([0, 1], [0, 1])

# Step 4: Visualize the constructed circuit
# 'mpl' uses matplotlib to draw a nice, colorful diagram
qc.draw('mpl')
```
