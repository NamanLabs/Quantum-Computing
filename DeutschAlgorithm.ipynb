# Lab 5: Implementation of Deutsch algorithm  

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit import transpile

qubits = QuantumRegister(2, 'q')
cbits = ClassicalRegister(1, 'c')
circuit = QuantumCircuit(qubits, cbits)

q0 = qubits[0]
q1 = qubits[1]

circuit.x(q1)
circuit.h(q0)
circuit.h(q1)

# Balanced Oracle
circuit.cx(q0, q1)

circuit.h(q0)

circuit.measure(q0, cbits[0])   # ⚠️ VERY IMPORTANT

simulator = AerSimulator()

compiled = transpile(circuit, simulator)

result = simulator.run(compiled, shots=1024).result()

counts = result.get_counts(compiled)   # safer form
print(counts)
```

In the Deutsch 1-bit algorithm:

Output 0 → Function is constant

Output 1 → Function is balanced
