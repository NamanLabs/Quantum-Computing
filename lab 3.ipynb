```python

# Importing necessary modules from Qiskit
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit

# Step 1: Create a Quantum Register with 2 qubits
# We name it 'q_reg' so it shows up clearly in the circuit diagram
q_reg = QuantumRegister(2, name='q_reg')

# Step 2: Create a Classical Register with 2 bits
# We name it 'c_reg' to store our final measurement results
c_reg = ClassicalRegister(2, name='c_reg')

# Step 3: Build the Quantum Circuit by combining both registers
qc = QuantumCircuit(q_reg, c_reg)

# Step 4: Apply basic gates to see the registers in action
qc.x(q_reg[0])  # Apply Pauli-X (NOT) gate to the first qubit
qc.h(q_reg[1])  # Apply Hadamard gate to the second qubit

# Step 5: Measure the Quantum Register into the Classical Register
# This maps q_reg[0] to c_reg[0] and q_reg[1] to c_reg[1]
qc.measure(q_reg, c_reg)

# Step 6: Visualize the circuit
qc.draw('mpl')
```
