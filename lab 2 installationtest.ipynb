# Lab 0: Installation and Setup of Qiskit on Windows Using Miniconda

# Both Anaconda Distribution and Miniconda include conda, but they differ in what’s pre-installed.
# Anaconda Distribution is a comprehensive installer optimized for data science and AI workflows with hundreds of packages automatically included, plus Anaconda Navigator.
# Miniconda is a lightweight version that includes only conda, Python, and their dependencies.

# To display all available conda environments on the system.

```python
! conda env list
```

# Creates a new isolated conda environment named cwq.

```python
! conda create --name cwq
```

# Activates the cwq environment for package installation and usage.

```python
! conda activate "D:\Codes\Global Venv\venv"
```

# Installs the pip package manager inside the active conda environment.

```python
! conda install pip
```

# Installs the core Qiskit quantum computing framework.

```python
! pip install qiskit
```

# Installs Matplotlib for visualization of quantum circuits and results.

```python
! pip install matplotlib
```

# Installs IBM Runtime module for accessing IBM Quantum cloud backends.

```python
! pip install qiskit_ibm_runtime
```

# Installs LaTeX encoding support for proper circuit diagram rendering.

```python
! pip install pylatexenc
```
